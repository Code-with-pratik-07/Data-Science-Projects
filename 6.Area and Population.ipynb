```python
import pandas as pd
cities=pd.read_csv("california_cities.csv")
print(cities.head())
```

```python
print(cities.isnull().sum())
```

```python
latitude,longitude,population,area=cities["latd"],cities["longd"],cities["population_total"],cities["area_total_sq_mi"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.gca().set_aspect('equal', adjustable='datalim')
plt.xlabel('Longitude')
plt.ylabel('Latiitude')
plt.colorbar()
for area in [100,300,500]:
    plt.scatter([],[],c="k",alpha=0.3,s=area,label=str(area)+ 'km$^2$')
plt.legend(scatterpoints=1,title="City areas")
plt.title("Area and Population of california cities")
plt.show()
```
